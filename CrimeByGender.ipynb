# Question 2: Victims of Violent Crimes by gender identity in the past 5 years of data

## 2015 data set

```python
Gender2015 = pd.read_excel("victims_sex_by_offense_category_2015.xls")
```

```python
Gender2015.head()
```

### Lets take out a little of the stuff we dont need. (The first 3 columns)

```python
Gender2015A = Gender2015[3:24]
```

```python
Gender2015A.head()
```

### Now lets flip this so we can see it more clearly

```python
Gender2015B = Gender2015A.T
Gender2015B.head()
```

### Ok now for renaming the columns

```python
Gender2015B.columns = ["Gender", "Total", "CrimesAgainstPersons", "AssaultOffenses", "HomicideOffenses", "HumanTraffickingOffenses", "Kidnapping/Abduction", "SexOffenses", "SexOffensesNonForcible", "CrimesAgainstProperty", "Bribery", "Burglary/BreakingEntering", "Counterfeiting/Forgery", "Destruction/Damage/Vandalism", "Embezzelment", "Extortion/Blackmail", "FraudOffenses", "Larceny/TheftOffenses", "MotorVehicleTheft", "Robbery", "StolenPropertyOffenses"]
```

```python
Gender2015B.head()
```

### And now to trim it down to just the numbers... 

```python
Gender2015C = Gender2015B[["Total", "CrimesAgainstPersons", "AssaultOffenses", "HomicideOffenses", "HumanTraffickingOffenses", "Kidnapping/Abduction", "SexOffenses", "SexOffensesNonForcible", "CrimesAgainstProperty", "Bribery", "Burglary/BreakingEntering", "Counterfeiting/Forgery", "Destruction/Damage/Vandalism", "Embezzelment", "Extortion/Blackmail", "FraudOffenses", "Larceny/TheftOffenses", "MotorVehicleTheft", "Robbery", "StolenPropertyOffenses"]]
```

```python
Gender2015D = Gender2015C[1:4]
```

```python
Gender2015D.head()
```

### At this point it is important to note Row 1 is Totals, Row 2 is male, row 3 is female and row 4 is unknown

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Now lets look at this with a pairplot and see what shows up... 

```python
sns.pairplot(Gender2015D)
```
